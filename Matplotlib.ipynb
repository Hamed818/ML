```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = { "A" : 30, "B" : 24, "AB" : 8, "O" : 38}
people_n = list(data.values())
blood_type = list(data.keys())

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

plt.bar(blood_type, people_n, color ='turquoise', width = 0.5)

plt.xlabel("blood type")
plt.ylabel("people number")
plt.title("blood type abundance")
```

```python
data = { "A" : 30, "B" : 24, "AB" : 8, "O" : 38}
people_n = list(data.values())
blood_type = list(data.keys())

fig = plt.figure(figsize=(8,8))
explode = (0.1, 0.0, 0.4, 0.0)
colors = ( "orange", "cyan", "indigo", "lightcoral")

plt.pie(people_n, labels = blood_type, explode= explode, shadow=True, colors=colors, frame=True,)
plt.show()


```
