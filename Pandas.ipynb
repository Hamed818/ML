##  First Method

```python
import pandas as pd

def First_Three (phone_num):
    return (phone_num)[:3]

H_B_df["District_Code"] = H_B_df["phone-number"].apply(First_Three)

Most_Frequent_District = H_B_df["District_Code"].value_counts()[:3]
Most_Frequent_District
```

## Second Method

```python
def first_three (phone_num):
    return (phone_num.split("-"))[0]

H_B_df["District_Code2"] = H_B_df["phone-number"].apply(first_three)

Most_Frequent_District = H_B_df["District_Code2"].value_counts()[:3]
Most_Frequent_District
```
